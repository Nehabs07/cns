```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Region': ['North America'] * 100 + ['Europe'] * 100 + ['Asia'] * 100,
    'Satisfaction': [5, 4, 3, 2, 1] * 20 + [4, 3, 3, 4, 2] * 20 + [3, 3, 3, 4, 2] * 20
}

df = pd.DataFrame(data)

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a count plot
plt.figure(figsize=(12, 6))
count_plot = sns.countplot(x='Satisfaction', hue='Region', data=df, palette='viridis')
count_plot.set_title('Distribution of Satisfaction Survey Responses Across Regions', fontsize=16)
count_plot.set_xlabel('Satisfaction Level', fontsize=14)
count_plot.set_ylabel('Count', fontsize=14)
plt.show()

# Create a box plot
plt.figure(figsize=(12, 6))
box_plot = sns.boxplot(x='Region', y='Satisfaction', data=df, palette='viridis')
box_plot.set_title('Blend of Satisfaction Survey Responses Across Regions', fontsize=16)
box_plot.set_xlabel('Region', fontsize=14)
box_plot.set_ylabel('Satisfaction Level', fontsize=14)
plt.show()
```

```python
#4. What r distribution and trend of responses to a statisfaction survey across diff regions.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Region': ['North', 'North', 'North', 'South', 'South', 'South', 'East', 'East', 'East', 'West', 'West', 'West'],
    'Satisfaction Score': [7, 8, 6, 5, 6, 7, 9, 8, 8, 5, 4, 6],
    'Survey Date': pd.to_datetime([
        '2024-01-01', '2024-02-01', '2024-03-01',
        '2024-01-01', '2024-02-01', '2024-03-01',
        '2024-01-01', '2024-02-01', '2024-03-01',
        '2024-01-01', '2024-02-01', '2024-03-01'
    ])
})

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a box plot for the distribution of satisfaction scores by region
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
sns.boxplot(x='Region', y='Satisfaction Score', data=data, palette='muted')
plt.title('Distribution of Satisfaction Scores by Region')
plt.xlabel('Region')
plt.ylabel('Satisfaction Score')

# Create a line plot for the trend of satisfaction scores over time by region
plt.subplot(1, 2, 2)
sns.lineplot(x='Survey Date', y='Satisfaction Score', hue='Region', data=data, marker='o')
plt.title('Trend of Satisfaction Scores Over Time by Region')
plt.xlabel('Survey Date')
plt.ylabel('Satisfaction Score')
plt.xticks(rotation=45)

# Display the plots
plt.tight_layout()
plt.show()
```
